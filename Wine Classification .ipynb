```python
X=df.drop('Cultivator',axis=1)
y=df['Cultivator']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, BaggingClassifier, AdaBoostClassifier , GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
```

## Using GridSearchCV to get the best params for the Algorithms
## As we have more than 2 categories, we use f1_weighted as the metric for evaluating model performance as it takes into consideration both precision and recall

### Finding best params for KNN

```python
from sklearn.model_selection import GridSearchCV
knn = KNeighborsClassifier()
param = {'n_neighbors':np.arange(1,50),'weights':['uniform','distance']}
GS=GridSearchCV(knn,param,cv=3,scoring='f1_weighted')
GS.fit(X_std,y)
```

```python
print('The best parameters for KNN are')
GS.best_params_
```

### Finding best params for Decision Tree

```python
dt = DecisionTreeClassifier(random_state=0)
```

```python
param = {'max_depth':np.arange(3,50),'criterion':['entropy','gini'],'min_samples_leaf':np.arange(3,20)}
GS=GridSearchCV(dt,param,scoring='f1_weighted')
GS.fit(X_std,y)
```

min_samples_leaf balances for max_depth
Here max_depth it's better to give after seeing the depth of fully grown DT

```python
print('The best parameters for Decision Tree are')
GS.best_params_
```

### The best Params for Random Forest and Bagging models are found manually as GridSearch CV gives the result on the basis of highest f1_weighted while we want to find the best model based on variance

### Finding best value of n_estimator for Random Forest

```python
RF_var = []
for val in np.arange(1,50):
    RF = RandomForestClassifier(criterion='entropy',n_estimators=val,random_state=0)
    kfold = model_selection.KFold(shuffle=True,n_splits=3,random_state=0)
    cv_results = model_selection.cross_val_score(RF,X_std,y,cv=kfold,scoring='f1_weighted')
    RF_var.append(np.var(cv_results,ddof=1))
print('We get the lowest variance error for n_estimators =',np.argmin(RF_var)+1)
```

### Finding best value of n_estimator for Bagging model 

```python
Bag_var = []
for val in np.arange(1,100):
    Bag = BaggingClassifier(n_estimators=val,random_state=0)
    kfold = model_selection.KFold(shuffle=True,n_splits=3,random_state=0)
    cv_results = model_selection.cross_val_score(Bag,X_std,y,cv=kfold,scoring='f1_weighted')
    Bag_var.append(np.var(cv_results,ddof=1))
print('We get the lowest variance error for n_estimators =',np.argmin(Bag_var)+1)
```

### Finding best value of n_estimator for AdaBoost model 

```python
Ad_bias = []
for val in np.arange(1,100):
    Ad = AdaBoostClassifier(n_estimators=val,random_state=0)
    kfold = model_selection.KFold(shuffle=True,n_splits=3,random_state=0)
    cv_results = model_selection.cross_val_score(Ad,X_std,y,cv=kfold,scoring='f1_weighted')
    Ad_bias.append(1-np.mean(cv_results))
print('We get the lowest bias error for n_estimators =',np.argmin(Ad_bias)+1)
    
```

### Finding best value of n_estimator for GradientBoost model 

```python
bias = []
for val in np.arange(1,100):
    GB = GradientBoostingClassifier(n_estimators=val,random_state=0)
    kfold = model_selection.KFold(shuffle=True,n_splits=3,random_state=0)
    cv_results = model_selection.cross_val_score(GB,X_std,y,cv=kfold,scoring='f1_weighted')
    bias.append(1-np.mean(cv_results))
print('We get the lowest bias error for n_estimators =',np.argmin(bias)+1)
    
```

```python
LR=LogisticRegression()
NB=GaussianNB()
KNN=KNeighborsClassifier(n_neighbors=25,weights='uniform')
DT=DecisionTreeClassifier(criterion='entropy', max_depth=3,min_samples_leaf=3)
RF=RandomForestClassifier(criterion='entropy',n_estimators=4)#Here using entropy as criterion as we got the best criterion for DT as Entropy
Bagged=BaggingClassifier(n_estimators=36,random_state=0)
AdaBoost=AdaBoostClassifier(n_estimators=15,random_state=0)
GBoost=GradientBoostingClassifier(n_estimators=24)
```

```python
models=[]
models.append(('LogR',LR))
models.append(('NaiveB',NB))
models.append(('KNN',KNN))
models.append(('DecisionTree',DT))
models.append(('RandomForest',RF))
models.append(('BaggingClassifier',Bagged))
models.append(('AdaBoost',AdaBoost))
models.append(('GBoost',GBoost))
```

```python
from sklearn import model_selection
results=[]
names=[]
for name,model in models:
    kfold = model_selection.KFold(shuffle=True, n_splits=5, random_state=0)
    cv_results = model_selection.cross_val_score(model,X_std,y,cv=kfold,scoring='f1_weighted')
    results.append(np.sqrt(np.abs(cv_results)))
    names.append(name)
    print('%s: %f (%f)'% (name,np.mean(np.sqrt(np.abs(cv_results))),np.std(np.sqrt(np.abs(cv_results)),ddof=1)))

#boxplot algorithm comparison
fig=plt.figure()
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

### Now Let's create a Bagged model of Logistic Regression to so that we get better variance

```python
BagLR_var = []
for val in np.arange(1,100):
    Bag = BaggingClassifier(base_estimator=LR,n_estimators=val,random_state=0)
    kfold = model_selection.KFold(shuffle=True,n_splits=3,random_state=0)
    cv_results = model_selection.cross_val_score(Bag,X_std,y,cv=kfold,scoring='f1_weighted')
    BagLR_var.append(np.var(cv_results,ddof=1))
print('We get the lowest variance error for n_estimators =',np.argmin(BagLR_var)+1)
```

#### Here we see that the Bagged model is not able to improve the variance for the LR model

### Let us now check by using Voting Classifier

```python
models.append(('StackedLRandNB',stackedLRNB))
models.append(('StackedLRandBag',stackedLRBag))
```

```python
models.append(('StackedLRNBandBag',stackedLRNBBag))
```
